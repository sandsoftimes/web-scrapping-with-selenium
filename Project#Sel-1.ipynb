```python
from selenium import webdriver #importing selenium library
from selenium.webdriver.common.by import By #This will import By that will be used in our code
import pandas as pd


web="https://www.audible.com/search" #Defining the website we want to scrape
driver=webdriver.Chrome() #Defining the webdriver i.e the opening of browser
driver.get(web) #We are getting the data of website on driver
driver.maximize_window() #By this the driver window will open all over the screen

container=driver.find_element(By.CLASS_NAME,'adbl-impression-container ') #copy the name from the inspect on the site and make the container here
# container.find_element('xpath','./li') #. represents the current /li represents li in current. Here the current is container and we are searching in container/li #Now we are going to locate the specific content of our need inside of this container
# container.find_elements('xpath','./li') #This will give multiple elements of li, above one was giving just the first one li only
products=container.find_elements('xpath','./li') # Store this in the products variable
#We want to scrape title,author name and length/runtime of audio book
#We are going to use contain function
#inspect the title, author and length again and find their origin tags
#Btw products is a list of li because find_elements give list
book_title=[] #Created empty list to store title
book_author=[] #Created empty list to store author name
book_length=[] #Created empty list to store book_length

for product in products:
    #Now use xpath in the inspect section
    #Use this xpath //h3[contains(@class,"bc-heading")]
    book_title.append(product.find_element('xpath','//h3[contains(@class,"bc-heading")]').text) #Take this from XPATH
    print(book_title)
    book_author.append(product.find_element('xpath','//li[contains(@class,"authorLabel")]').text) #Take this from XPATH
    book_length.append(product.find_element('xpath','//li[contains(@class,"runtimeLabel")]').text) #Take this from XPATH
    
# driver.quit()

#Now we want to create dataframe to export data in json or csv file

df_books=pd.DataFrame({'title':book_title,'author':book_author,'length':book_length})
df_books.to_csv('books.csv',index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

web = "https://www.audible.com/search"
# path = '/Users/frank/Downloads/chromedriver'
driver = webdriver.Chrome()
driver.get(web)
driver.maximize_window()

# Locating the box that contains all the audiobooks listed in the page
container = driver.find_element(By.CLASS_NAME,'adbl-impression-container ')
# Getting all the audiobooks listed (the "/" gives immediate child nodes)
products = container.find_elements('xpath','./li')
# Initializing storage
book_title = []
book_author = []
book_length = []
# Looping through the products list (each "product" is an audiobook)
for product in products:
    # We use "contains" to search for web elements that contain a particular text, so we avoid building long XPATH
    title=product.find_element('xpath','.//h3[contains(@class, "bc-heading")]').text
    book_title.append(title)  # Storing data in list
    book_author.append(product.find_element('xpath','.//li[contains(@class, "authorLabel")]').text)
    book_length.append(product.find_element('xpath','.//li[contains(@class, "runtimeLabel")]').text)

driver.quit()
# Storing the data into a DataFrame and exporting to a csv file
df_books = pd.DataFrame({'title': book_title, 'author': book_author, 'length': book_length})
df_books.to_csv('books.csv', index=False)
```
